```python
import pandas as pd
import numpy as np
```

# Read in health data

```python
df_health = pd.read_csv('./Data/health.csv')
```

```python
df_health.head()
```

# Filter only New York City

```python
df_ny = df_health[df_health['StateAbbr'] == 'NY']
```

```python
np.unique(df_ny['PlaceName'])
```

```python
df_nyc = df_ny[df_ny['PlaceName'] == 'New York']
df_nyc.shape
```

```python
df_nyc.head()
```

```python
df_nyc = df_nyc.reset_index(drop=True)
df_nyc.head()
```

```python
df_nyc.to_csv('./Data/NYC_Health.csv', index=False)
```

```python
df_nyc_subset = df_nyc[['PlaceFIPS', 'TractFIPS', 'Place_TractID',
                'DIABETES_CrudePrev', 'HIGHCHOL_CrudePrev', 'OBESITY_CrudePrev']]
df_nyc_subset.head()
```

```python
df_nyc_subset.to_csv('./Data/NYC_Health_Features.csv', index=False)
```
